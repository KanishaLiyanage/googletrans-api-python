```python
import numpy as np
import pandas as pd
from googletrans import LANGUAGES
from googletrans import Translator

translator = Translator(service_urls=['translate.googleapis.com'])
```

```python
test_csv = pd.read_csv('source_csv/test.csv')
```

```python
test_csv.head()
```

```python
new_test = test_csv.copy()
```

```python
new_test.head(10)
```

```python
translated_title = []
translated_description = []
```

```python
for index, row in new_test.iterrows():
    translated_title.append(translator.translate(row['Title'], dest='si').text)
    translated_description.append(translator.translate(row['Description'], dest='si').text)
```

```python
new_test = new_test.copy()
new_test.loc[:, 'Translated Title'] = translated_title
new_test.loc[:, 'Translated Description'] = translated_description
```

```python
new_test.to_csv('translated_csv/sinhala_corpus.csv', encoding='utf-8')
```
